```python
import matplotlib.pyplot as plt
import numpy as np
import services

DENSITY = 1000

STEP1 = 2
STEP2 = 2
STEP3 = 0.1

STEP = STEP1

FILE1 = "set_1.csv"
FILE2 = "set_2.csv"
FILE3 = "set_3.csv"

FILE = FILE1

def sort(distribution):
    return np.sort(distribution)

sorted_starting_distribution = sort(services.read_csv(FILE))

size = len(sorted_starting_distribution)
expected_value = services.get_expected_value(sorted_starting_distribution)
variance = services.get_variance(sorted_starting_distribution, expected_value)
standart_deviation = variance ** 0.5

services.draw_distribution_plot(sorted_starting_distribution, DENSITY, 'График функции распределения для файла {}'.format(FILE))
```

```python
services.draw_hist(sorted_starting_distribution, STEP, 'График гистограммы для данных из файла {}'.format(FILE))
```

```python
p = services.get_geometric_p_by_expected_value(expected_value)
if p > 0 and p < 1:
    services.draw_hist(sort(np.random.geometric(p, size)), STEP, 'График гистограммы геометрического распределения по мат ожиданию из файла {}'.format(FILE))
```

```python
p = services.get_binomial_p_by_expected_value(expected_value, size)
services.draw_hist(sort(np.random.binomial(size, p, size)), STEP, 'График гистограммы биномиального распределения по мат ожиданию из файла {}'.format(FILE))
```

```python
services.draw_hist(sort(np.random.poisson(expected_value, size)), STEP, 'График гистограммы распределения Пуассона из файла {}'.format(FILE))
```

```python
low, high = services.get_uniform_low_high(expected_value, variance)
services.draw_hist(sort(np.random.uniform(low, high, size)), STEP, 'График гистограммы равномерного распределения из файла {}'.format(FILE))
```

```python
services.draw_hist(sort(np.random.normal(expected_value, standart_deviation, size)), STEP, 'График гистограммы нормального распределения из файла {}'.format(FILE))
```

```python
services.draw_hist(sort(np.random.exponential(expected_value, size)), STEP, 'График гистограммы экспоненциального распределения из файла {}'.format(FILE))
```
