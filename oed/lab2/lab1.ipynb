```python
import matplotlib.pyplot as plt
import numpy as np
import math
import lab1lib as ll


DENSITY = 1000

VALS_PART = 0.005
WIDTH_PART = 0.3
CUT_COEFF = 1

WINDOW1 = 1
WINDOW2 = 100
WINDOW3 = 50

WINDOW = WINDOW3

STEP1 = 1
STEP2 = 2
STEP3 = 0.05

STEP = STEP3

FILE1 = "set_1.csv"
FILE2 = "set_2.csv"
FILE3 = "set_3.csv"

FILE = FILE3


def clear(distribution):
    return ll.clear(distribution, VALS_PART, WIDTH_PART, CUT_COEFF)
    #return distribution

def sort(distribution):
    return np.sort(distribution)


sorted_starting_distribution = ll.approximate(clear(sort(ll.read_csv(FILE))), WINDOW)

size = len(sorted_starting_distribution)
expected_value = ll.get_expected_value(sorted_starting_distribution)
variance = ll.get_variance(sorted_starting_distribution, expected_value)
standart_deviation = variance ** 0.5


generated = ll.generate_distributions(sorted_starting_distribution)
name, starting_distribution = ll.get_distribution(generated, sorted_starting_distribution, 0.01)

plot_from = math.floor(sorted_starting_distribution[0])
plot_to = math.ceil(sorted_starting_distribution[-1])


if name:

    match name:

        case ll.StandartDistribution.BINOMIAL:
            print("Binomial")

        case ll.StandartDistribution.POISSON:
            print("Poisson")

        case ll.StandartDistribution.GEOMETRIC:
            print("Geometric")

        case ll.StandartDistribution.NORMAL:
            print("Normal")

        case ll.StandartDistribution.UNIFORM:
            print("Uniform")
        
        case ll.StandartDistribution.EXPONENTIAL:
            print("Exponential")
        
    ll.draw_plot(generated[name], DENSITY, plot_from, plot_to)

ll.draw_plot(starting_distribution, DENSITY, plot_from, plot_to)
```
